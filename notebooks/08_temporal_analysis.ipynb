# 08 - Temporal Analysis: Monthly Binning

This notebook performs a comprehensive temporal analysis with monthly aggregations to understand:

1. **Equity Evolution**: Start/end equity, monthly returns, cumulative performance
2. **Leverage & Exposure**: AUM utilization, gross exposure, capital deployment
3. **Risk Metrics**: Monthly volatility, drawdowns, VaR, regime-conditioned risk
4. **Asset Selection**: Products traded, concentration (HHI), diversification trends
5. **Volume & Activity**: Number of trades, turnover, trade duration, capital committed
6. **Regime Analysis**: Performance by regime, regime transitions, selection patterns

**Key structural break**: Dec 26, 2023 (Phase 1 Ã¢â€ â€™ Phase 2 transition)

---

### Important: Equity Fields Semantics

| Field | Description | Use Case |
|-------|-------------|----------|
| `equity_EOD` | Actual account equity at end of day | Account size display, leverage calculation |
| `equity_normalized` | Adjusted equity masking capital injections/withdrawals | **Performance/return calculations** |
| `AUM` | `equity_EOD Ãƒâ€” 10` (leverage capacity) | Utilization rate |
| `total_invested_amount_EOD` | Capital committed in open positions | Exposure analysis |

**Critical**: For reliable performance metrics, we use `equity_normalized` to avoid distortions from the ~16x capital injection on Dec 26, 2023.

## 1.1 Benchmark Matrix Analysis - Reconstructing Equity from Prices Ãƒâ€” Weights

This section loads the benchmark_matrix.xlsx file containing daily asset amounts and percentages,
matches each asset with its last available price, and calculates the benchmark equity as:

**Benchmark Equity = ÃŽÂ£ (asset_price_i Ãƒâ€” asset_weight_i)**

```python
# Load benchmark matrix from Excel - has two sheets: 'assets' and 'percentage'
benchmark_matrix_path = DATA_RAW / 'benchmark' / 'benchmark_matrix.xlsx'

# Read Excel file - inspect sheets
xl = pd.ExcelFile(benchmark_matrix_path)
print(f"Sheet names: {xl.sheet_names}")

# Load both sheets
benchmark_assets_raw = pd.read_excel(benchmark_matrix_path, sheet_name='assets')
benchmark_pct_raw = pd.read_excel(benchmark_matrix_path, sheet_name='percentage')

print(f"\n--- Assets sheet (absolute values) ---")
print(f"Shape: {benchmark_assets_raw.shape}")
print(f"Columns: {benchmark_assets_raw.columns[:5].tolist()}...")

print(f"\n--- Percentage sheet (weights) ---")
print(f"Shape: {benchmark_pct_raw.shape}")
print(f"Columns: {benchmark_pct_raw.columns[:5].tolist()}...")
```

```python
# Process both sheets: Column A = asset names, Column B = text, Columns C+ = dates

def process_benchmark_sheet(df_raw: pd.DataFrame, sheet_name: str) -> pd.DataFrame:
    """
    Process a benchmark matrix sheet.
    Structure: Column A = asset names, Column B = description, Columns C+ = dates as headers
    Returns: DataFrame with dates as index, assets as columns
    """
    print(f"\nProcessing '{sheet_name}' sheet...")
    
    # Identify text columns vs date columns
    text_cols = []
    date_cols = []
    for col in df_raw.columns:
        if isinstance(col, (pd.Timestamp, np.datetime64)) or 'datetime' in str(type(col)):
            date_cols.append(col)
        else:
            text_cols.append(col)
    
    print(f"  Text columns: {text_cols}")
    print(f"  Date columns: {len(date_cols)}")
    
    # Use first text column as asset identifier
    asset_col = text_cols[0]
    print(f"  Using '{asset_col}' as asset identifier")
    
    # Set asset names as index
    df_clean = df_raw.set_index(asset_col)
    
    # Drop other text columns
    cols_to_drop = [c for c in text_cols[1:] if c in df_clean.columns]
    if cols_to_drop:
        print(f"  Dropping: {cols_to_drop}")
        df_clean = df_clean.drop(columns=cols_to_drop)
    
    # Transpose: dates -> index, assets -> columns
    df_final = df_clean.T
    df_final.index = pd.to_datetime(df_final.index)
    df_final.index.name = 'date'
    df_final.columns = df_final.columns.astype(str)
    
    # Convert to numeric, coerce errors
    df_final = df_final.apply(pd.to_numeric, errors='coerce')
    
    print(f"  Final shape: {df_final.shape}")
    print(f"  Date range: {df_final.index.min()} to {df_final.index.max()}")
    
    return df_final

# Process both sheets
benchmark_assets = process_benchmark_sheet(benchmark_assets_raw, 'assets')
benchmark_pct = process_benchmark_sheet(benchmark_pct_raw, 'percentage')

print(f"\n--- Summary ---")
print(f"Assets matrix: {benchmark_assets.shape} (absolute amounts)")
print(f"Percentage matrix: {benchmark_pct.shape} (weights)")
print(f"\nSample asset names: {benchmark_assets.columns[:10].tolist()}")

# Show sample data
print(f"\n--- Sample Assets Data (first 3 dates, first 5 assets) ---")
print(benchmark_assets.iloc[:3, :5])

print(f"\n--- Sample Percentage Data (first 3 dates, first 5 assets) ---")
print(benchmark_pct.iloc[:3, :5])
```

```python
# Load asset prices from algorithm CSV files
# Each CSV has OHLC data with datetime, open, high, low, close

def load_asset_prices(algo_folder: Path, asset_ids: list) -> pd.DataFrame:
    """
    Load the last daily close price for each asset.
    Each algo CSV has 4-hourly OHLC data. We extract the last close per day.
    
    Returns: DataFrame with dates as index and asset_id as columns, containing close prices.
    """
    all_prices = {}
    missing_assets = []
    
    for asset_id in asset_ids:
        csv_path = algo_folder / f"{asset_id}.csv"
        if not csv_path.exists():
            missing_assets.append(asset_id)
            continue
        
        try:
            # Load the CSV
            df = pd.read_csv(csv_path)
            
            # Parse datetime
            if 'datetime' in df.columns:
                df['datetime'] = pd.to_datetime(df['datetime'])
                df['date'] = df['datetime'].dt.date
            elif 'date' in df.columns:
                df['date'] = pd.to_datetime(df['date']).dt.date
            else:
                # Try first column
                df['datetime'] = pd.to_datetime(df.iloc[:, 0])
                df['date'] = df['datetime'].dt.date
            
            # Get last close price per day
            daily_close = df.groupby('date')['close'].last()
            daily_close.index = pd.to_datetime(daily_close.index)
            
            all_prices[asset_id] = daily_close
            
        except Exception as e:
            print(f"Error loading {asset_id}: {e}")
            missing_assets.append(asset_id)
    
    if missing_assets:
        print(f"\nMissing/failed assets: {len(missing_assets)}")
        if len(missing_assets) <= 10:
            print(f"  {missing_assets}")
    
    # Combine all prices into a single DataFrame
    prices_df = pd.DataFrame(all_prices)
    prices_df = prices_df.sort_index()
    
    return prices_df

# Get unique asset IDs from the assets matrix
asset_ids = benchmark_assets.columns.tolist()
print(f"Loading prices for {len(asset_ids)} assets...")

# Load prices
algo_folder = DATA_RAW / 'algorithms'
asset_prices = load_asset_prices(algo_folder, asset_ids)

print(f"\nLoaded prices shape: {asset_prices.shape}")
print(f"Date range: {asset_prices.index.min()} to {asset_prices.index.max()}")
print(f"\nSample prices (first 5 assets, last 5 days):")
asset_prices.iloc[-5:, :5]
```

```python
# Calculate Benchmark Index (starts at 100, changes based on weighted average returns)
#
# Index calculation:
#   - Day 0: Index = 100
#   - Day N: Index = Index_{N-1} × (1 + weighted_return_N)
#   - weighted_return = Σ (weight_i × return_i)
#   - return_i = (price_i_today - price_i_yesterday) / price_i_yesterday

def calculate_benchmark_index(weights_matrix: pd.DataFrame, 
                               prices_matrix: pd.DataFrame,
                               base_value: float = 100.0) -> tuple:
    """
    Calculate benchmark index like a weighted index (e.g., S&P 500).
    
    Parameters:
    -----------
    weights_matrix : DataFrame - percentage weights per asset per day (should sum to ~100 or ~1)
    prices_matrix : DataFrame - close prices per asset per day
    base_value : float - starting index value (default 100)
    
    Returns:
    --------
    tuple: (result_df, weights_normalized, asset_returns)
        - result_df: DataFrame with index value and related metrics
        - weights_normalized: DataFrame with normalized weights (sum to 1)
        - asset_returns: DataFrame with daily returns per asset
    """
    # Align dates
    common_dates = weights_matrix.index.intersection(prices_matrix.index)
    common_dates = common_dates.sort_values()
    print(f"Common dates: {len(common_dates)}")
    
    if len(common_dates) == 0:
        print("WARNING: No common dates!")
        return pd.DataFrame(), pd.DataFrame(), pd.DataFrame()
    
    # Align assets
    common_assets = weights_matrix.columns.intersection(prices_matrix.columns)
    print(f"Common assets: {len(common_assets)} / {len(weights_matrix.columns)}")
    
    weights = weights_matrix.loc[common_dates, common_assets].copy()
    prices = prices_matrix.loc[common_dates, common_assets].copy()
    
    # Forward fill prices for missing values
    prices = prices.ffill()
    
    # Fill NaN weights with 0
    weights = weights.fillna(0)
    
    # Normalize weights to sum to 1 (if they sum to ~100, divide by 100)
    weight_sums = weights.sum(axis=1)
    print(f"Weight sums - mean: {weight_sums.mean():.2f}, min: {weight_sums.min():.2f}, max: {weight_sums.max():.2f}")
    
    if weight_sums.mean() > 50:  # Weights are in percentage (0-100)
        print("Normalizing weights from 0-100 to 0-1")
        weights = weights / 100
        weight_sums = weights.sum(axis=1)
    
    # Normalize so weights sum to 1 each day
    weights_normalized = weights.div(weight_sums, axis=0).fillna(0)
    
    # Calculate daily returns for each asset
    asset_returns = prices.pct_change()
    
    # Calculate weighted portfolio return each day
    # weighted_return = Σ (weight_i × return_i)
    weighted_returns = (weights_normalized * asset_returns).sum(axis=1)
    
    # Handle first day (no return)
    weighted_returns.iloc[0] = 0
    
    # Calculate cumulative index value
    # Index_t = base_value × Π(1 + r_i) for i = 1 to t
    index_values = base_value * (1 + weighted_returns).cumprod()
    
    # Build result DataFrame
    result = pd.DataFrame({
        'index_value': index_values,
        'daily_return_pct': weighted_returns * 100,
        'cumulative_return_pct': (index_values / base_value - 1) * 100,
        'n_assets': (weights > 0).sum(axis=1),
        'total_weight_pct': weight_sums * 100,
    })
    
    # Add contribution breakdown (which assets contributed most)
    # Handle rows with all NA by filling with empty string/0
    contributions = weights_normalized * asset_returns * 100  # In percentage points
    
    # For idxmax/idxmin, we need to handle rows with all NA values
    # Use a mask to identify valid rows (not all NA)
    valid_rows = ~contributions.isna().all(axis=1)
    
    result['top_contributor'] = ''
    result['top_contribution_pct'] = 0.0
    result['bottom_contributor'] = ''
    result['bottom_contribution_pct'] = 0.0
    
    if valid_rows.any():
        result.loc[valid_rows, 'top_contributor'] = contributions.loc[valid_rows].idxmax(axis=1)
        result.loc[valid_rows, 'top_contribution_pct'] = contributions.loc[valid_rows].max(axis=1)
        result.loc[valid_rows, 'bottom_contributor'] = contributions.loc[valid_rows].idxmin(axis=1)
        result.loc[valid_rows, 'bottom_contribution_pct'] = contributions.loc[valid_rows].min(axis=1)
    
    return result, weights_normalized, asset_returns

print("Calculating benchmark index (base = 100)...")
print("Formula: Index_t = Index_{t-1} × (1 + Σ(weight_i × return_i))")
```

```python
# Execute the index calculation
index_result, weights_norm, asset_returns = calculate_benchmark_index(
    weights_matrix=benchmark_pct,
    prices_matrix=asset_prices, 
    base_value=100.0
)

print(f"\nIndex result shape: {index_result.shape}")
print(f"\n--- Index Statistics ---")
print(f"Start value: {index_result['index_value'].iloc[0]:.2f}")
print(f"End value: {index_result['index_value'].iloc[-1]:.2f}")
print(f"Total return: {index_result['cumulative_return_pct'].iloc[-1]:.2f}%")
print(f"Avg daily return: {index_result['daily_return_pct'].mean():.4f}%")
print(f"Daily volatility: {index_result['daily_return_pct'].std():.4f}%")
print(f"Annual volatility: {index_result['daily_return_pct'].std() * np.sqrt(252):.2f}%")

# Sharpe ratio (assuming 0 risk-free rate)
sharpe = index_result['daily_return_pct'].mean() / index_result['daily_return_pct'].std() * np.sqrt(252)
print(f"Sharpe ratio (annualized): {sharpe:.2f}")

print(f"\n--- First 5 days ---")
print(index_result.head())

print(f"\n--- Last 5 days ---")
print(index_result.tail())
```

## 2. Monthly Aggregation Functions

```python
def compute_monthly_metrics(trades_df: pd.DataFrame) -> pd.DataFrame:
    """
    Compute monthly aggregated metrics from trade-level data.
    
    IMPORTANT: Uses equity_normalized for return calculations to mask capital flows.
    Uses equity_EOD for actual account size and leverage calculations.
    
    Returns DataFrame with monthly bins containing:
    - Equity metrics (both EOD and normalized)
    - Leverage/exposure metrics
    - Volume/activity metrics
    - Asset selection metrics
    """
    df = trades_df.copy()
    df['month'] = df['dateOpen'].dt.to_period('M')
    
    monthly = []
    
    for month, group in df.groupby('month'):
        # Get last/first day data for equity/AUM snapshots
        last_trade = group.loc[group['dateOpen'].idxmax()]
        first_trade = group.loc[group['dateOpen'].idxmin()]
        
        # Equity EOD metrics (actual account size)
        equity_eod_start = first_trade.get('equity_EOD', np.nan)
        equity_eod_end = last_trade.get('equity_EOD', np.nan)
        
        # Equity NORMALIZED metrics (for performance - masks capital flows)
        equity_norm_start = first_trade.get('equity_normalized', np.nan)
        equity_norm_end = last_trade.get('equity_normalized', np.nan)
        
        # Monthly return from NORMALIZED equity (correct for capital flows)
        if equity_norm_start > 0 and not np.isnan(equity_norm_start):
            monthly_return_pct = (equity_norm_end / equity_norm_start - 1) * 100
        else:
            monthly_return_pct = np.nan
        
        # AUM and leverage (use EOD for actual account metrics)
        aum_avg = group['AUM'].mean()
        aum_end = last_trade.get('AUM', np.nan)
        invested_avg = group['total_invested_amount_EOD'].mean()
        invested_max = group['total_invested_amount_EOD'].max()
        
        # Leverage = invested / equity_EOD (implicit leverage)
        equity_eod_mean = group['equity_EOD'].mean()
        leverage_avg = invested_avg / equity_eod_mean if equity_eod_mean > 0 else np.nan
        
        # Utilization = invested / AUM
        utilization_avg = invested_avg / aum_avg if aum_avg > 0 else np.nan
        utilization_max = invested_max / aum_avg if aum_avg > 0 else np.nan
        
        # Volume/Activity metrics
        n_trades = len(group)
        total_volume = group['volume'].sum()
        avg_trade_size = group['volume'].mean()
        avg_trade_duration = group['trade_duration_days'].mean()
        median_trade_duration = group['trade_duration_days'].median()
        
        # Asset selection metrics
        unique_products = group['productname'].nunique()
        product_counts = group['productname'].value_counts()
        
        # HHI concentration (based on trade count per product)
        trade_shares = product_counts / product_counts.sum()
        hhi_trades = (trade_shares ** 2).sum()
        
        # Top 5 concentration
        top5_share = product_counts.head(5).sum() / product_counts.sum()
        top_product = product_counts.index[0] if len(product_counts) > 0 else None
        top_product_share = product_counts.iloc[0] / product_counts.sum() if len(product_counts) > 0 else 0
        
        monthly.append({
            'month': month,
            'month_start': month.start_time,
            # Equity EOD (actual account size)
            'equity_eod_start': equity_eod_start,
            'equity_eod_end': equity_eod_end,
            'equity_eod_change_pct': (equity_eod_end / equity_eod_start - 1) * 100 if equity_eod_start > 0 else np.nan,
            # Equity Normalized (for returns - masks capital flows)
            'equity_norm_start': equity_norm_start,
            'equity_norm_end': equity_norm_end,
            'monthly_return_norm_pct': monthly_return_pct,  # KEY: return from normalized equity
            # AUM & Leverage
            'aum_avg': aum_avg,
            'aum_end': aum_end,
            'invested_avg': invested_avg,
            'invested_max': invested_max,
            'leverage_avg': leverage_avg,
            'utilization_avg_pct': utilization_avg * 100 if utilization_avg else np.nan,
            'utilization_max_pct': utilization_max * 100 if utilization_max else np.nan,
            # Volume/Activity
            'n_trades': n_trades,
            'total_volume': total_volume,
            'avg_trade_size': avg_trade_size,
            'avg_trade_duration_days': avg_trade_duration,
            'median_trade_duration_days': median_trade_duration,
            # Asset selection
            'n_unique_products': unique_products,
            'hhi_trades': hhi_trades,
            'top5_concentration_pct': top5_share * 100,
            'top_product': top_product,
            'top_product_share_pct': top_product_share * 100,
        })
    
    return pd.DataFrame(monthly)
```

```python
def compute_monthly_returns_from_weights(algo_returns: pd.DataFrame, 
                                          benchmark_weights: pd.DataFrame) -> pd.DataFrame:
    """
    Compute monthly portfolio returns using daily algo returns and weights.
    """
    # Align data
    common_idx = algo_returns.index.intersection(benchmark_weights.index)
    returns = algo_returns.loc[common_idx]
    weights = benchmark_weights.loc[common_idx]
    
    # Ensure same columns
    common_cols = returns.columns.intersection(weights.columns)
    returns = returns[common_cols]
    weights = weights[common_cols]
    
    # Daily portfolio return = sum(weight_i * return_i)
    daily_portfolio_return = (returns * weights).sum(axis=1)
    
    # Monthly aggregation
    monthly_returns = daily_portfolio_return.resample('MS').apply(
        lambda x: (1 + x).prod() - 1
    )
    
    # Also compute monthly volatility (annualized from daily)
    monthly_vol = daily_portfolio_return.resample('MS').std() * np.sqrt(252)
    
    # Monthly Sharpe (assuming 0 risk-free rate)
    monthly_sharpe = monthly_returns / (daily_portfolio_return.resample('MS').std() + 1e-10)
    
    return pd.DataFrame({
        'monthly_return': monthly_returns,
        'monthly_return_pct': monthly_returns * 100,
        'monthly_vol_ann': monthly_vol,
        'monthly_sharpe': monthly_sharpe,
        'n_trading_days': daily_portfolio_return.resample('MS').count(),
    })
```

```python
def compute_monthly_risk_metrics(algo_returns: pd.DataFrame,
                                  benchmark_weights: pd.DataFrame) -> pd.DataFrame:
    """
    Compute monthly risk metrics: drawdown, VaR, CVaR, etc.
    """
    # Align data
    common_idx = algo_returns.index.intersection(benchmark_weights.index)
    returns = algo_returns.loc[common_idx]
    weights = benchmark_weights.loc[common_idx]
    common_cols = returns.columns.intersection(weights.columns)
    
    # Daily portfolio return
    daily_portfolio_return = (returns[common_cols] * weights[common_cols]).sum(axis=1)
    
    # Cumulative return for drawdown calculation
    cumulative = (1 + daily_portfolio_return).cumprod()
    running_max = cumulative.expanding().max()
    drawdown = (cumulative / running_max - 1)
    
    monthly_risk = []
    
    for month, group in daily_portfolio_return.groupby(pd.Grouper(freq='MS')):
        if len(group) == 0:
            continue
            
        # Drawdown for this month
        month_dd = drawdown.loc[group.index]
        
        # VaR and CVaR (95%)
        var_95 = np.percentile(group, 5)  # 5th percentile = 95% VaR
        cvar_95 = group[group <= var_95].mean() if len(group[group <= var_95]) > 0 else var_95
        
        # Skewness and kurtosis
        skewness = group.skew()
        kurtosis = group.kurtosis()
        
        # Positive/negative days
        pct_positive = (group > 0).mean() * 100
        
        monthly_risk.append({
            'month_start': month,
            'max_drawdown_pct': month_dd.min() * 100,
            'avg_drawdown_pct': month_dd.mean() * 100,
            'var_95_daily': var_95 * 100,
            'cvar_95_daily': cvar_95 * 100,
            'skewness': skewness,
            'kurtosis': kurtosis,
            'pct_positive_days': pct_positive,
            'worst_day_pct': group.min() * 100,
            'best_day_pct': group.max() * 100,
        })
    
    return pd.DataFrame(monthly_risk).set_index('month_start')
```

```python
def compute_monthly_asset_selection(benchmark_weights: pd.DataFrame) -> pd.DataFrame:
    """
    Analyze asset selection patterns monthly from weights data.
    """
    monthly_selection = []
    
    for month, group in benchmark_weights.groupby(pd.Grouper(freq='MS')):
        if len(group) == 0:
            continue
        
        # Average weights across the month
        avg_weights = group.mean()
        
        # Count active products (weight > threshold)
        threshold = 0.001  # 0.1% minimum weight to be considered "active"
        active_products = (avg_weights > threshold).sum()
        
        # Total exposure
        total_exposure = avg_weights.sum()
        
        # Concentration metrics
        nonzero_weights = avg_weights[avg_weights > 0]
        if len(nonzero_weights) > 0:
            weight_shares = nonzero_weights / nonzero_weights.sum()
            hhi = (weight_shares ** 2).sum()
            top5_weight = nonzero_weights.nlargest(5).sum() / nonzero_weights.sum()
            top10_weight = nonzero_weights.nlargest(10).sum() / nonzero_weights.sum()
            max_weight = nonzero_weights.max()
            min_weight = nonzero_weights.min()
            top_product = nonzero_weights.idxmax()
        else:
            hhi = 1.0
            top5_weight = 1.0
            top10_weight = 1.0
            max_weight = 0
            min_weight = 0
            top_product = None
        
        # Weight dispersion
        weight_std = nonzero_weights.std() if len(nonzero_weights) > 1 else 0
        
        monthly_selection.append({
            'month_start': month,
            'n_active_products': active_products,
            'total_exposure': total_exposure,
            'hhi_weights': hhi,
            'top5_concentration_pct': top5_weight * 100,
            'top10_concentration_pct': top10_weight * 100,
            'max_weight_pct': max_weight * 100,
            'min_weight_pct': min_weight * 100,
            'weight_std': weight_std,
            'top_product': top_product,
            'effective_n': 1 / hhi if hhi > 0 else 0,  # Effective number of products
        })
    
    return pd.DataFrame(monthly_selection).set_index('month_start')
```

```python
def compute_monthly_regime_stats(regime_labels: pd.DataFrame) -> pd.DataFrame:
    """
    Compute regime statistics per month.
    """
    # Ensure datetime index
    if not isinstance(regime_labels.index, pd.DatetimeIndex):
        regime_labels.index = pd.to_datetime(regime_labels.index)
    
    regime_col = regime_labels.columns[0]  # Get the regime column name
    
    monthly_regime = []
    
    for month, group in regime_labels.groupby(pd.Grouper(freq='MS')):
        if len(group) == 0:
            continue
        
        regimes = group[regime_col]
        
        # Dominant regime
        dominant = regimes.mode().iloc[0] if len(regimes.mode()) > 0 else np.nan
        dominant_pct = (regimes == dominant).mean() * 100
        
        # Regime distribution
        regime_counts = regimes.value_counts(normalize=True) * 100
        
        # Regime transitions
        transitions = (regimes != regimes.shift()).sum() - 1  # Subtract 1 for first day
        
        monthly_regime.append({
            'month_start': month,
            'dominant_regime': dominant,
            'dominant_pct': dominant_pct,
            'n_transitions': max(0, transitions),
            'regime_0_pct': regime_counts.get(0, 0),
            'regime_1_pct': regime_counts.get(1, 0),
            'regime_2_pct': regime_counts.get(2, 0),
            'regime_3_pct': regime_counts.get(3, 0),
            'n_days': len(regimes),
        })
    
    return pd.DataFrame(monthly_regime).set_index('month_start')
```
