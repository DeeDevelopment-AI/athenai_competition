# Phase 3: Baseline Portfolio Strategies Analysis

This notebook analyzes the results of backtesting classical portfolio allocation strategies
(MPT baselines) against the benchmark.

## Strategies Tested
1. **Equal Weight** - 1/N allocation across all algorithms
2. **Risk Parity** - Weights inversely proportional to volatility
3. **Min Variance** - Quadratic optimization minimizing portfolio variance
4. **Max Sharpe** - Optimization maximizing Sharpe ratio
5. **Momentum** - Overweight algorithms with best recent momentum
6. **Vol Targeting** - Scale exposure to maintain target volatility

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from pathlib import Path

# Set style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('Set2')

# Load trial data
output_dir = Path('data/reports')
df = pd.read_csv(output_dir / 'phase3_trials.csv')

print(f'Total trials: {len(df)}')
print(f'Baselines: {df["config_baseline_name"].unique().tolist()}')
```

## 1. Trial Summary by Baseline

```python
# Separate main trials from walk-forward trials
main_trials = df[df['notes'].isna() | (df['notes'] == '')].copy()

# Summary statistics
summary = main_trials.groupby('config_baseline_name').agg({
    'sharpe_val': ['count', 'mean', 'std', 'max'],
    'sharpe_test': ['mean', 'std', 'max'],
    'return_ann_val': 'mean',
    'max_dd_val': 'mean',
}).round(4)

summary.columns = ['_'.join(col).strip() for col in summary.columns.values]
summary = summary.sort_values('sharpe_val_mean', ascending=False)
print('Summary by Baseline (sorted by validation Sharpe):')
display(summary)
```

## 2. Sharpe Ratio Comparison

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Box plot of validation Sharpe
order = main_trials.groupby('config_baseline_name')['sharpe_val'].mean().sort_values(ascending=False).index
sns.boxplot(data=main_trials, x='config_baseline_name', y='sharpe_val', order=order, ax=axes[0])
axes[0].axhline(y=0.95, color='r', linestyle='--', label='Benchmark Sharpe')
axes[0].set_title('Validation Sharpe by Baseline')
axes[0].set_xlabel('Baseline')
axes[0].set_ylabel('Sharpe Ratio')
axes[0].tick_params(axis='x', rotation=45)
axes[0].legend()

# Box plot of test Sharpe
sns.boxplot(data=main_trials, x='config_baseline_name', y='sharpe_test', order=order, ax=axes[1])
axes[1].axhline(y=0.95, color='r', linestyle='--', label='Benchmark Sharpe')
axes[1].set_title('Test Sharpe by Baseline')
axes[1].set_xlabel('Baseline')
axes[1].set_ylabel('Sharpe Ratio')
axes[1].tick_params(axis='x', rotation=45)
axes[1].legend()

plt.tight_layout()
plt.savefig(output_dir / 'sharpe_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3. Train vs Validation vs Test Sharpe

```python
# Scatter plot of train vs test Sharpe
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for baseline in main_trials['config_baseline_name'].unique():
    subset = main_trials[main_trials['config_baseline_name'] == baseline]
    axes[0].scatter(subset['sharpe_train'], subset['sharpe_val'], label=baseline, alpha=0.7, s=50)
    axes[1].scatter(subset['sharpe_val'], subset['sharpe_test'], label=baseline, alpha=0.7, s=50)

# Add diagonal line
for ax in axes:
    lims = [min(ax.get_xlim()[0], ax.get_ylim()[0]), max(ax.get_xlim()[1], ax.get_ylim()[1])]
    ax.plot(lims, lims, 'k--', alpha=0.5, label='y=x')

axes[0].set_xlabel('Train Sharpe')
axes[0].set_ylabel('Validation Sharpe')
axes[0].set_title('Train vs Validation Sharpe')
axes[0].legend(loc='upper left', fontsize=8)

axes[1].set_xlabel('Validation Sharpe')
axes[1].set_ylabel('Test Sharpe')
axes[1].set_title('Validation vs Test Sharpe')
axes[1].legend(loc='upper left', fontsize=8)

plt.tight_layout()
plt.savefig(output_dir / 'train_val_test_scatter.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Lookback Window Impact

```python
# Filter baselines that use lookback
lookback_baselines = main_trials[main_trials['config_lookback_window'] > 0].copy()

fig, ax = plt.subplots(figsize=(10, 6))

for baseline in lookback_baselines['config_baseline_name'].unique():
    subset = lookback_baselines[lookback_baselines['config_baseline_name'] == baseline]
    grouped = subset.groupby('config_lookback_window')['sharpe_val'].mean()
    ax.plot(grouped.index, grouped.values, marker='o', label=baseline, linewidth=2)

ax.set_xlabel('Lookback Window (days)')
ax.set_ylabel('Mean Validation Sharpe')
ax.set_title('Impact of Lookback Window on Performance')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(output_dir / 'lookback_impact.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Walk-Forward Validation Results

```python
# Get walk-forward trials
wf_trials = df[df['notes'].fillna('').str.contains('Walk-forward')].copy()

if len(wf_trials) > 0:
    # Extract fold number
    wf_trials['fold'] = wf_trials['notes'].str.extract(r'fold (\d+)').astype(float)
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Test Sharpe across folds
    for baseline in wf_trials['config_baseline_name'].unique():
        subset = wf_trials[wf_trials['config_baseline_name'] == baseline].sort_values('fold')
        axes[0].plot(subset['fold'], subset['sharpe_test'], marker='o', label=baseline, linewidth=2)
    
    axes[0].axhline(y=0, color='k', linestyle='--', alpha=0.3)
    axes[0].axhline(y=0.95, color='r', linestyle='--', alpha=0.5, label='Benchmark')
    axes[0].set_xlabel('Walk-Forward Fold')
    axes[0].set_ylabel('Test Sharpe')
    axes[0].set_title('Walk-Forward Test Sharpe by Fold')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    # Distribution of test Sharpe
    sns.boxplot(data=wf_trials, x='config_baseline_name', y='sharpe_test', ax=axes[1])
    axes[1].axhline(y=0.95, color='r', linestyle='--', label='Benchmark')
    axes[1].set_xlabel('Baseline')
    axes[1].set_ylabel('Test Sharpe')
    axes[1].set_title('Walk-Forward Test Sharpe Distribution')
    axes[1].tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.savefig(output_dir / 'walk_forward_results.png', dpi=150, bbox_inches='tight')
    plt.show()
else:
    print('No walk-forward trials found')
```

## 6. Best Configuration Details

```python
# Best config per baseline
best_configs = []
for baseline in main_trials['config_baseline_name'].unique():
    subset = main_trials[main_trials['config_baseline_name'] == baseline]
    best_idx = subset['sharpe_val'].idxmax()
    best = subset.loc[best_idx]
    best_configs.append({
        'Baseline': baseline,
        'Lookback': best['config_lookback_window'],
        'Rebalance': best['config_rebalance_frequency'],
        'Shrinkage': best.get('config_shrinkage', 'N/A'),
        'Train Sharpe': f"{best['sharpe_train']:.2f}",
        'Val Sharpe': f"{best['sharpe_val']:.2f}",
        'Test Sharpe': f"{best['sharpe_test']:.2f}",
    })

best_df = pd.DataFrame(best_configs)
best_df = best_df.sort_values('Val Sharpe', ascending=False)
display(best_df)
```

## 7. Benchmark Comparison Summary

```python
# Load benchmark metrics
bench_metrics = json.load(open('data/phase2/benchmark_metrics.json'))

print('='*60)
print('BENCHMARK METRICS')
print('='*60)
print(f'Sharpe Ratio:     {bench_metrics["sharpe_ratio"]:.4f}')
print(f'Ann. Return:      {bench_metrics["annualized_return"]:.2%}')
print(f'Ann. Volatility:  {bench_metrics["annualized_volatility"]:.2%}')
print(f'Max Drawdown:     {bench_metrics["max_drawdown"]:.2%}')
print(f'Sortino Ratio:    {bench_metrics["sortino_ratio"]:.4f}')
print(f'Calmar Ratio:     {bench_metrics["calmar_ratio"]:.4f}')

print('\n' + '='*60)
print('BASELINES vs BENCHMARK')
print('='*60)

for _, row in best_df.iterrows():
    improvement = (float(row['Val Sharpe']) / bench_metrics['sharpe_ratio'] - 1) * 100
    print(f"{row['Baseline']:15s}: Val Sharpe = {row['Val Sharpe']:>5s} ({improvement:+.0f}% vs benchmark)")
```

## 8. Key Insights and Recommendations

### Key Findings

1. **Risk Parity dominates**: Simple inverse-volatility weighting outperforms complex optimization
2. **Equal Weight is robust**: Hard to beat consistently, especially in walk-forward
3. **Momentum underperforms**: Suggests algorithms may exhibit mean-reversion rather than momentum
4. **All baselines beat benchmark**: 2-3x improvement in Sharpe ratio

### Recommendations for RL (Phase 4)

1. **Baseline target**: RL must beat Risk Parity's Sharpe of ~3.4
2. **Stability focus**: Reduce walk-forward variance while maintaining high average
3. **Regime awareness**: High variance in Risk Parity suggests regime-dependent allocation opportunity
4. **Feature selection**: 21-day volatility is critical; include in RL state

```python
# Save key results
results_summary = {
    'best_baseline': 'risk_parity',
    'best_val_sharpe': float(best_df.iloc[0]['Val Sharpe']),
    'benchmark_sharpe': bench_metrics['sharpe_ratio'],
    'improvement_vs_benchmark': float(best_df.iloc[0]['Val Sharpe']) / bench_metrics['sharpe_ratio'],
    'recommended_config': {
        'baseline': 'risk_parity',
        'lookback': 21,
        'rebalance': 'monthly',
    },
    'rl_target_sharpe': 3.5,
}

with open(output_dir / 'phase3_key_results.json', 'w') as f:
    json.dump(results_summary, f, indent=2)

print('Key results saved to phase3_key_results.json')
print(json.dumps(results_summary, indent=2))
```
