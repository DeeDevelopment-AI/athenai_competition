````python
# Load data to verify incongruencies
algo_returns = pd.read_parquet(PROCESSED / 'algo_returns.parquet')
benchmark_weights = pd.read_parquet(PROCESSED / 'benchmark_weights.parquet')

# Normalize timezones
if algo_returns.index.tz is not None:
    algo_returns.index = algo_returns.index.tz_localize(None)
if benchmark_weights.index.tz is not None:
    benchmark_weights.index = benchmark_weights.index.tz_localize(None)

print(f"Algo returns shape: {algo_returns.shape}")
print(f"Benchmark weights shape: {benchmark_weights.shape}")
````

## 2.1 INCONGRUENCY: Benchmark Returns Calculation

### Problem:
Three different benchmark return sources exist:

| Source | Location | Ann. Return | Notes |
|--------|----------|-------------|-------|
| Loaded | benchmark_performance.csv | 2.59% | "Ground truth" |
| Derived (01) | weights Ã— algo_returns | 4.61% | Simple calculation |
| Derived (06) | aligned weights Ã— returns | ~4.4% | With index alignment |

### Correlation:
- Loaded vs Derived: **0.72** (should be >0.95)

### Root Causes:
1. Weight timing: weights may be end-of-day while returns are intraday
2. Missing costs: ~2% annual drag from transaction costs
3. Execution gaps: slippage, partial fills

````python
# Verify the benchmark return discrepancy
common_cols = algo_returns.columns.intersection(benchmark_weights.columns)
common_idx = algo_returns.index.intersection(benchmark_weights.index)

algo_aligned = algo_returns.loc[common_idx, common_cols]
weights_aligned = benchmark_weights.loc[common_idx, common_cols]

# Derived returns
derived_returns = (algo_aligned * weights_aligned).sum(axis=1)

# Try to load original benchmark returns if available
try:
    from src.data.loader import DataLoader
    loader = DataLoader('../data/raw/')
    benchmark_raw = loader.load_benchmark()
    loaded_returns = benchmark_raw.returns
    
    # Align
    if hasattr(loaded_returns.index, 'tz') and loaded_returns.index.tz is not None:
        loaded_returns.index = loaded_returns.index.tz_localize(None)
    
    common_dates = derived_returns.index.intersection(loaded_returns.index)
    
    corr = derived_returns.loc[common_dates].corr(loaded_returns.loc[common_dates])
    
    print(f"Derived returns: {derived_returns.mean() * 252:.2%} annualized")
    print(f"Loaded returns:  {loaded_returns.loc[common_dates].mean() * 252:.2%} annualized")
    print(f"Correlation:     {corr:.3f}")
    print(f"\nGap: {(derived_returns.mean() - loaded_returns.loc[common_dates].mean()) * 252:.2%} annual")
except Exception as e:
    print(f"Could not load benchmark returns: {e}")
    print(f"Derived returns: {derived_returns.mean() * 252:.2%} annualized")
````

### SOLUTION:

1. **For backtesting**: Use derived returns (weights Ã— algo_returns) but apply a **cost adjustment**:
   ```python
   adjusted_returns = derived_returns - daily_cost_estimate
   # where daily_cost_estimate â‰ˆ 2%/252 â‰ˆ 0.008% per day
   ```

2. **For validation**: Use loaded returns as ground truth

3. **For RL training**: Use derived returns (so agent can learn the relationship)

## 2.2 INCONGRUENCY: Regime Detection Approaches

### Problem:
Three different regime detection methodologies:

| Notebook | Input | Method | Philosophy |
|----------|-------|--------|------------|
| 04 | Benchmark returns | Investment Clock (4 regimes) | Predict macro cycle |
| 05 | Market proxy (avg all algos) | Fuzzy + period analysis | Find performance patterns |
| 06 | Benchmark + universe features | Latent inference (HMM) | Explain benchmark rotation |

### Issue:
Each produces **different regime labels** for the same dates!

````python
# Compare regime labels from different notebooks
regime_files = {
    'Investment Clock (04)': 'regime_labels.parquet',
    'Latent Inference (06)': 'latent_regime_labels.parquet',
}

regimes = {}
for name, filename in regime_files.items():
    try:
        df = pd.read_parquet(PROCESSED / filename)
        regimes[name] = df.iloc[:, 0]  # First column
        print(f"{name}: {len(df)} rows, {df.iloc[:, 0].nunique()} unique regimes")
    except FileNotFoundError:
        print(f"{name}: File not found")

# Check concordance if both exist
if len(regimes) == 2:
    r1, r2 = list(regimes.values())
    common = r1.index.intersection(r2.index)
    
    # Can't directly compare since labels differ
    print(f"\nCommon dates: {len(common)}")
    print(f"\n{list(regimes.keys())[0]} distribution:")
    print(r1.value_counts(normalize=True).round(3))
    print(f"\n{list(regimes.keys())[1]} distribution:")
    print(r2.value_counts(normalize=True).round(3))
````

### SOLUTION:

These are **complementary approaches**, not conflicting:

1. **Investment Clock (04)**: Use for macro context and interpretability
2. **Latent Inference (06)**: Use for benchmark rotation explanation

**Recommendation**: Use **Latent Inference** for the meta-allocator because:
- It directly explains benchmark behavior
- It provides P(family | regime) for policy learning
- It doesn't assume we can predict macro cycles

## 2.4 INCONGRUENCY: Activity Handling

### Problem:
Different notebooks handle inactive algorithms differently:

| Notebook | Method | Issue |
|----------|--------|-------|
| 01-05 | 0 = inactive | Can't distinguish 0 return from inactive |
| 06 | ActivityMask | Explicit boolean mask |

### Impact:
- Statistics calculated on "all data" include periods when algo wasn't trading
- Cross-sectional correlations include zeros
- Family aggregates may be biased

````python
# Demonstrate the activity handling difference
sample_algo = algo_returns.columns[0]
sample_returns = algo_returns[sample_algo]

# Old approach: count non-zero
old_active = (sample_returns != 0).sum()

# New approach: explicit mask (non-NaN AND non-zero)
new_active = ((~sample_returns.isna()) & (sample_returns != 0)).sum()

print(f"Algorithm: {sample_algo}")
print(f"Total rows: {len(sample_returns)}")
print(f"Active (old method - non-zero): {old_active}")
print(f"Active (new method - non-NaN & non-zero): {new_active}")
print(f"NaN count: {sample_returns.isna().sum()}")
print(f"Zero count: {(sample_returns == 0).sum()}")
````

### SOLUTION:

**Adopt ActivityMask from notebook 06** across all analysis:
